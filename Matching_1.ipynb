Матчинг через Jaro без оптимізації
Результат показав, що без додаткової обробки текстів, це мало що дасть, адже явно не-метчі мають схожість 0.96-0.97 тільки через подібність організації компанії(наприклад Організація з обмеженою відповідальністю - в даному випадку назва з 5-10 символів вже майже не вирішує)

```python
with open('data\\opensanctions_result.json', 'w') as g:
    for file_name in os.listdir('data'):
        if('.csv' in file_name):
            print(file_name)
            dfs = pd.read_csv('data\\' + file_name, sep = ';', chunksize = 1000)
            for df in dfs:
                print(f'File has: {len(df)} record')
        
                for row in df.iterrows():
                    with open('data\\opensanctions_processed.json', 'r') as f:
                        for line in f:
                            obj = json.loads(line)
        
                            #print(obj['caption'])
                            jarowinkler = list()
                            if(isinstance(obj['name'], list)):
                                if(len(obj['name']) > 0):
                                    for name in obj['name']:
                                        jarowinkler.append(distance.get_jaro_distance(name, row[1].FULL_NAME))
                                    max_jarowinkler = max(jarowinkler)
                                else:
                                    max_jarowinkler = distance.get_jaro_distance(obj['caption'], row[1].FULL_NAME)
                            else:
                                max_jarowinkler = distance.get_jaro_distance(obj['caption'], row[1].FULL_NAME)
                            if(max_jarowinkler >= 0.97):
                                print(max_jarowinkler)
                                print(obj['caption'])
                                print(obj['name'])
                                print(row[1].FULL_NAME)
                                print('\n\n')
        
                                obj['similarity'] = max_jarowinkler
                                obj['FULL_NAME'] = row[1].FULL_NAME
                                obj['TIN'] = row[1].TIN
        
                                g.write(json.dumps(obj) + '\n')
```
