```python
from langdetect import detect
import json
import os
```

Зрозумів, що даних надто багато, тому вирішив відкинути дані, які явно не заматчаться з неякими
В даному випадку на основі мови(усі записи в csv - українською)
Краще було б перевіряти по полю country в json'і, але я його не зберіг + можна розглядати це як попередню спробу на простіших даних і якби все було ок, то можна було окремо розглянути дані, які записані англійською
При цьому замінюю список імен різними мовами однією(в пріорітеті українська)

```python
counter_general = 0
counter_uk = 0
counter_ru = 0
counter_problem = 0

with open('data\\opensanctions_processed.json', 'r') as f:
    with open('data\\opensanctions_processed_uk.json', 'w') as u:
        with open('data\\opensanctions_processed_ru.json', 'w') as r:
            with open('data\\opensanctions_processed_problem.json', 'w') as p:
                for line in f:
                    try:
                        counter_general += 1
                        obj = json.loads(line)
                        languages = dict()
                        if(len(obj['name']) > 0):
                            for name in obj['name']:
                                languages[detect(name)] = name
                        languages[detect(obj['caption'])] = obj['caption']
                            
                        if('uk' in languages.keys()):
                            obj['name'] = languages['uk']
                            u.write(json.dumps(obj) + '\n')
                            counter_uk += 1
                        elif('ru' in languages.keys()):
                            obj['name'] = languages['ru']
                            r.write(json.dumps(obj) + '\n')
                            counter_ru += 1
                    except:
                        p.write(json.dumps(obj) + '\n')
                        counter_problem += 1
            print('General: ', counter_general)
            print('UK: :', counter_uk)
            print('RU: :', counter_ru)
            print('Issues: :', counter_problem)
```

```python
print('Difference only-UK: ', counter_general - counter_uk)
print('Difference UK+RU: ', counter_general - counter_uk - counter_ru)
```
