Отримання пар ознак батьківський-дочірній

```python
def get_dicts_recursive(field, field_name : str, parent_name : str):
    result = list()
    if(isinstance(field, dict)):
        for key in field.keys():
            result = result + get_dicts_recursive(field[key], key, field_name)
    if(isinstance(field, list)):
        for elem in field:
            if(isinstance(field, dict)):
                for key in field.keys():
                    result = result + get_dicts_recursive(field[key], key, field_name)
    result.append(f'{parent_name}-{field_name}')
    return result
```

```python
with open('data\\opensanctions_nonbinary.json', 'r', encoding = 'utf8') as f:
    schemas = list()
    result = list()
    
    for line in f:
        obj = json.loads(line)
        result = result + get_dicts_recursive(obj, 'parent', 'parent')
        result = list(dict.fromkeys(result))
```

Коментар: Поле [sanctions] є тільки в [properties]
Відповідно, якщо санкції накладені опосередковано(на власника компанії, а не саму компанію; чи на родича), то їх потрібно шукати в інших полях

Пошук в яких об'єктах зустрічаються санкції(перевірка чи в усіх)

```python
def get_parent_sanction_recurcive(field):
    result = list()
    if(isinstance(field, dict)):
        for key in field.keys():
            result = result + get_parent_sanction_recurcive(field[key])

        if('properties' in field.keys()):
            if('sanctions' in field['properties'].keys()):
                result.append(field['schema'])
    if(isinstance(field, list)):
        for elem in field:
            if(isinstance(field, dict)):
                for key in field.keys():
                    result = result + get_parent_sanction_recurcive(field[key])
    return result
```

```python
with open('data\\opensanctions_nonbinary.json', 'r', encoding = 'utf8') as f:
    schemas = list()
    result = list()
    
    for line in f:
        obj = json.loads(line)
        result = result + get_parent_sanction_recurcive(obj)
        result = list(dict.fromkeys(result))
```
